Create a dictionary of student names and their ages

```python
students = {"Alice": 20, "Bob": 22, "Charlie": 19}
print(students)
```

Add a new key-value pair

```python
d = {'a': 1, 'b': 2, 'c': 3}
d['d'] = 4
print(d)
```

Get all keys

```python
info = {'name': 'John', 'age': 25, 'city': 'New York'}
print(info.keys())
```

Get all values

```python
langs = {'python': 3, 'java': 2, 'c++': 1}
print(langs.values())
```

Check if key exists

```python
person = {'name': 'Alice', 'age': 30, 'city': 'London'}
print('age' in person)
```

Remove key

```python
d = {'a': 1, 'b': 2, 'temp': 3, 'c': 4}
d.pop('temp')
print(d)
```

 Sum of values

```python
marks = {'math': 85, 'science': 92, 'english': 78}
print(sum(marks.values()))
```

Squares dictionary

```python
squares = {x: x**2 for x in range(1, 6)}
print(squares)
```

Character frequency

```python
word = "hello"
freq = {}
for ch in word:
    freq[ch] = freq.get(ch, 0) + 1
print(freq)
```

 Merge dictionaries

```python
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
merged = {**d1, **d2}
print(merged)
```

 Access nested value


```python
nested = {'person': {'name': 'Alice', 'age': 25}}
print(nested['person']['name'])
```

 Dictionary with list values

```python
d = {'fruits': ['apple', 'banana'], 'colors': ['red', 'blue']}
print(d)
```

Add to list in dictionary

```python
d = {'fruits': ['apple', 'banana'], 'colors': ['red', 'blue']}
d['fruits'].append('orange')
print(d)
```

Dictionary with tuple values

```python
d = {'coordinates': (10, 20), 'rgb': (255, 0, 0)}
print(d)
```
