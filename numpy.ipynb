```python
# NumPy Arithmetic Functions Example (Single Cell)

import numpy as np

# Creating two arrays
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print("Array A:", a)
print("Array B:", b)

# Addition
print("\nAddition (np.add):")
print(np.add(a, b))

# Subtraction
print("\nSubtraction (np.subtract):")
print(np.subtract(a, b))

# Multiplication
print("\nMultiplication (np.multiply):")
print(np.multiply(a, b))

# Division
print("\nDivision (np.divide):")
print(np.divide(a, b))

# Power
print("\nPower (np.power):")
print(np.power(a, b))

# Modulus
print("\nModulus (np.mod):")
print(np.mod(a, b))

# Floor Division
print("\nFloor Division (np.floor_divide):")
print(np.floor_divide(a, b))

# Absolute Difference
print("\nAbsolute Difference (np.abs):")
print(np.abs(a - b))
```

```python
data= np.array([10,20,np.nan,40,50,np.nan,70])
data
```

```python
print("original data:", data)
```

```python
print("null position:",np.isnan(data))
```

```python
print("tot null values:",np.isnan(data).sum())
```

```python
clean_data=data[np.isnan(data)]
print("after removing null:", clean_data) 
data
```

```python
replace_zero=np.nan_to_num(data,nan=0)
print("replace nan with 0:", replace_zero)
```

```python
# 5. Replace Null Values with Mean
mean_value = np.nanmean(data)
replace_mean = np.where(np.isnan(data), mean_value, data)
print("Replace NaN with Mean:", replace_mean)
# 6. Replace Null Values with Median
median_value = np.nanmedian(data)
replace_median = np.where(np.isnan(data), median_value, data)
print("Replace NaN with Median:", replace_median)

# 7. Replace Null Values with Custom Value
replace_custom = np.where(np.isnan(data), 999, data)
print("Replace NaN with Custom Value:", replace_custom)

# 8. Ignore Null Values in Calculations
print("Mean ignoring NaN:", np.nanmean(data))
print("Sum ignoring NaN:", np.nansum(data))
print("Std ignoring NaN:", np.nanstd(data))
print("Min ignoring NaN:", np.nanmin(data))
print("Max ignoring NaN:", np.nanmax(data))
```
