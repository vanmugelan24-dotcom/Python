1: Create a set of first 10 natural numbers

```python
natural_numbers = set(range(1, 11))
print("Answer:", natural_numbers)
```

2: Add element 11 to set {1, 2, 3, 4, 5}

```python
s1 = {1, 2, 3, 4, 5}
s1.add(11)
print("Answer:", s1)
```

3: Remove element 3 from set {1, 2, 3, 4, 5, 6}

```python
s2 = {1, 2, 3, 4, 5, 6}
s2.remove(3)
print("Answer:", s2)
```

4: Find the intersection of {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}

```python
s3 = {1, 2, 3, 4, 5}
s4 = {4, 5, 6, 7, 8}
print("Answer:", s3.intersection(s4))
```

 5: Find the difference between {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}

```python
print("Answer:", s3.difference(s4))
```

6: Check if 5 is in set {1, 2, 3, 4, 6, 7, 8}

```python
s5 = {1, 2, 3, 4, 6, 7, 8}
print("Answer:", 5 in s5)
```

7: Find the length of set {'a', 'b', 'c', 'd', 'e'}

```python
s6 = {'a', 'b', 'c', 'd', 'e'}
print("Answer:", len(s6))
```

8: Create a set of vowels from string 'hello world

```python
vowels = {'a', 'e', 'i', 'o', 'u'}
string_set = set("hello world")
print("Answer:", string_set.intersection(vowels))
```

9: Remove duplicates from list [1, 2, 2, 3, 4, 4, 5, 6, 6, 7] using set

```python
lst = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
print("Answer:", set(lst))
```

10: Check if {1, 2, 3} is a subset of {1, 2, 3, 4, 5, 6}

```python
subset_check = {1, 2, 3}.issubset({1, 2, 3, 4, 5, 6})
print("Answer:", subset_check)
```
