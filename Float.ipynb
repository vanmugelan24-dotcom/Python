Calculate the average of 3.14, 2.718, 1.618, 0.577

```python
numbers = [3.14, 2.718, 1.618, 0.577]
average = sum(numbers) / len(numbers)
print("Average:", average)
```

Convert 98.6 Fahrenheit to Celsius (F = C * 9/5 + 32)

```python
F = 98.6
C = (F - 32) * 5/9
print("Celsius:", C)
```

Calculate the compound interest on $1000 at 5.5% for 3 years

```python
P = 1000
r = 0.055
t = 3
n = 1  # compounded annually

A = P * (1 + r/n)**(n*t)
CI = A - P

print("Compound Interest:", CI)
print("Total Amount:", A)
```

Find the hypotenuse of a right triangle with sides 3.5 and 4.2

```python
import math

a = 3.5
b = 4.2

c = math.sqrt(a**2 + b**2)
print("Hypotenuse:", c)
```

Calculate the volume of a sphere with radius 7.8

```python
import math

r = 7.8
V = (4/3) * math.pi * r**3
print("Volume of sphere:", V)
```

Calculate the simple interest: Principal=2500, Rate=6.5%, Time=2.5 years

```python
P = 2500       
R = 6.5        
T = 2.5        

SI = (P * R * T) / 100
Total = P + SI

print("Simple Interest:", SI)
print("Total Amount:", Total)
```
