 Find the most frequent character in "mississippi"

```python
from collections import Counter
text = "mississippi"
freq = Counter(text)
most_common = freq.most_common(1)[0]
print(most_common)
```

Count consonants in "Hello World"

```python
text = "Hello World"
vowels = "aeiouAEIOU"
consonants = [ch for ch in text if ch.isalpha() and ch not in vowels]
print(len(consonants))
```

Find the longest word in "Python is a programming language"

```python
text = "Python is a programming language"
words = text.split()
longest = max(words, key=len)
print(longest)
```

 Remove all punctuation from "Hello, World! How are you?"

```python
import string
text = "Hello, World! How are you?"
result = "".join([ch for ch in text if ch not in string.punctuation])
print(result)
```
