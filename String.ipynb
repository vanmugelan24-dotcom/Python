Reverse the string "Python Programming"

```python
text = "Python Programming"
print(text[::-1])
```

Check if "racecar" is a palindrome

```python
text = "racecar"

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

Count the number of words in "Python is a great programming language"

```python
text = "Python is a great programming language"
words = text.split()         
count = len(words)            
print("Number of words:", count)
```
