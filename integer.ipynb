```python
number = 12345
sum_of_digits = 0

for digit in str(number):
    sum_of_digits += int(digit)

print(sum_of_digits)
```

Check if 97 is a prime number

```python
number = 97
is_prime = True

if number <= 1:
    is_prime = False
else:
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break

if is_prime:
    print("97 is a prime number")
else:
    print("97 is not a prime number")
```

Find the factorial of 8

```python
n = 8
factorial = 1

for i in range(1, n + 1):
    factorial *= i

print(factorial)
```

Calculate the average of numbers: 15, 23, 31, 42, 56

```python
numbers = [15, 23, 31, 42, 56]
average = sum(numbers) / len(numbers)
print(average)
```

Find the greatest common divisor (GCD) of 48 and 36

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(48, 36))
```

Calculate the sum of first 20 odd numbers

```python
sum_of_odds = 0
number = 1 

for i in range(20):
    sum_of_odds += number
    number += 2 

print("Sum of first 20 odd numbers is:", sum_of_odds)
```
